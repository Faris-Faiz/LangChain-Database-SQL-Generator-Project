````python
client = OpenAI(api_key=OPENAI_API_KEY)

completion = client.chat.completions.create(
  model="gpt-4-0125-preview",
  messages=[
    {"role": "system", "content": f"Based on the table schema below, write a SQL query that would answer the user's question:{db_schema}"},
    {"role": "user", "content": f"Question: {question}\nSQL Query:"}
  ]
)
````

````python
import re
# Extract the content from the message
completion_message = completion.choices[0].message
content = completion_message.content
#print(content)

# Use regex to extract SQL code within triple backticks
sql_code_match = re.search(r'```sql\n(.*?)```', content, re.DOTALL)

if sql_code_match:
    sql_code = sql_code_match.group(1)
    print(sql_code)
else:
    print("SQL code not found in the completion message.")
````

````python
question_list = ["Retrieve all customers' names and emails.", "Find all tracks with a unit price greater than $0.99.", "List all invoices along with customer names.",
                 "Get all track names along with their album titles and artist names.", "Show the total amount of sales per country.",
                 "Find the name of the artist who has the highest number of tracks in the database.",
                 "List customers and their total spending, ordered by the highest spender first.",
                 "Rank customers by the number of purchases they've made.", "Find the most popular track of each genre.",
                 "List the top 3 selling tracks of each genre since 2020."]

sqlQuery_list = []
````

````python
def generateSQLStatement(question, db_schema):
  completion = client.chat.completions.create(
  model="gpt-4-0125-preview",
  messages=[
    {"role": "system", "content": f"Based on the table schema below, write a SQL query that would answer the user's question:{db_schema}"},
    {"role": "user", "content": f"Question: {question}\nSQL Query:"}
  ]
)

  # Extract the content from the message
  completion_message = completion.choices[0].message
  content = completion_message.content

  # Use regex to extract SQL code within triple backticks
  sql_code_match = re.search(r'```sql\n(.*?)```', content, re.DOTALL)

  if sql_code_match:
      sql_code = sql_code_match.group(1)
      return sql_code
  else:
      return "SQL code not found in the completion message."
````

````python
for question in question_list:
  SQL_Query = generateSQLStatement(question, db_schema)
  sqlQuery_list.append(SQL_Query)

len(sqlQuery_list)
````

````python
# Combining the list sqlQuery_list and questions_list into a dictionary for easier reference
# to convert lists to dictionary
res = {}
for key in question_list:
    for value in sqlQuery_list:
        res[key] = value
        sqlQuery_list.remove(value)
        break

# Printing resultant dictionary
print("Resultant dictionary is : " + str(res))
````

````python
# Dictionary to store formatted SQL queries
formatted_sql_queries = {}

# Iterate through the dictionary and replace the literal '\n' with actual newlines
for request, sql_query in res.items():
    formatted_sql_query = sql_query.replace("\\n", "\n")
    formatted_sql_queries[request] = formatted_sql_query
````
